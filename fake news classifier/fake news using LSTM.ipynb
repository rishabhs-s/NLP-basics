```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data=pd.read_csv("train.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data=data.dropna()#removing null values
```

```python
data.shape
```

```python
X=data.drop('label',axis=1)
```

```python
y=data['label']
```

```python
X #independent var
```

```python
y #dependent variable
```

```python
message=X.copy()
```

```python
message.reset_index(inplace=True)# resets index as null values are removed so we need to order index again
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
corpus=[]
```

```python
len(message)
```

```python
voc_size=8000
```

```python
for i in range(0,len(message)):    
    text = re.sub('[^a-zA-Z]', ' ', message['title'][i])
    text=text.lower()
    text=text.split()
    text=[ps.stem(word) for word in text if not word in stopwords.words("english")]

    text= " ".join(text)
    corpus.append(text)

    
    
    
```

```python
print(len(corpus))
```

```python
corpus
```

```python
from tensorflow.keras.preprocessing.text import one_hot
onehot_rep=[one_hot(words,voc_size)for words in corpus] 
onehot_rep
```

```python

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
sent_length=20  #max length of sentence
embedded=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
print(embedded)
```

```python
embedded[0]#message/sentence is converted to numbers with max 17 numbers /words are numerically represented and zero is padded if less than sent_length
```

```python
from tensorflow.keras.layers import Dropout
features=40 #vector of features
model=Sequential()
model.add(Embedding(voc_size,features,input_length=sent_length))#1st layer
model.add(Dropout(0.3))
model.add(LSTM(100))# LSTM layer with 100 nuerons
model.add(Dropout(0.3))
model.add(Dense(1,activation="sigmoid")) #dense layer for output
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=["accuracy"])
```

```python
model.summary()
```

```python
import numpy as np
X_final=np.array(embedded)
y_final=np.array(y)
```

```python
X_final.shape
```

```python
X_final
```

```python
y_final.shape
```

```python
len(embedded),y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.25
                                                   )
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred=model.predict_classes(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
