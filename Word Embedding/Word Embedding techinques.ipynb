```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
voc_size=10000 #size of dictionary
```

```python
one_hot=for words in sent:
```

```python
one_hot_rep=[one_hot(words,voc_size) for words in sent]
```

```python
one_hot_rep
```

```python
from tensorflow.keras.layers import Embedding #embedding layers
from tensorflow.keras.preprocessing.sequence import pad_sequences #padding to make all sentences of smae length
from tensorflow.keras.models import Sequential
```

```python
embedded=pad_sequences(one_hot_rep,padding="pre",maxlen=8)
```

```python
embedded
```
